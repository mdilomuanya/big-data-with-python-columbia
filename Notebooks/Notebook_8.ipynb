```python
%matplotlib ipympl
import matplotlib.pylab as plt
import numpy as np
from scipy import stats
import pandas as pd
```

```python
# Reading the CSV file 'Brian.csv' into a pandas
brian = pd.read_csv('Brian.csv', header=None)

# Converting the DataFrame to a NumPy array
brian2d = brian.values

# Adding first part of exclaimation point
brian2d[2:12, 29] = 4

# Adding second part of exclaimation point
brian2d[14:15, 29] = 4

# Creating a figure and axes for the plot
fig, ax = plt.subplots()

# Setting the figure size to 4 inches by 4 inches
fig.set_size_inches(4, 4)

# Displaying the 2D array as an image and adding color
cax = ax.imshow(brian2d, cmap='gist_earth_r', interpolation='none')

plt.show()
```

```python
# Read the Excel file into pandas
q3 = pd.read_excel('MD_Notebook8_Q3.xlsx')

# Convert the DataFrame into a NumPy array
q3_2d = q3.values

# Create a figure and axis for plotting
fig, ax = plt.subplots()

# Set the figure size
fig.set_size_inches(4, 4)

# Plot the three y-value columns against the x-values, adding labels for each line
ax.plot(q3_2d[:, 0], q3_2d[:, 1], label="Y1 - Linear")  # First column (X) vs. second column (Y1)
ax.plot(q3_2d[:, 0], q3_2d[:, 2], label="Y2 - Quadratic")  # X vs. third column (Y2)
ax.plot(q3_2d[:, 0], q3_2d[:, 3], label="Y3 - Sine Wave")  # X vs. fourth column (Y3)

# Add labels for the axes
ax.set_xlabel("X-Values") 
ax.set_ylabel("Y-Values") 

# Add a title to the plot
ax.set_title("Plot of Three Functions")

# Add a legend
ax.legend()

# Display the plot
plt.show()
```
