```python
%matplotlib ipympl
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

## Question 1

```python
df=pd.read_csv('well_data.csv')

# Create the scatter plot
fig, ax = plt.subplots(figsize=(8, 6))
ax.scatter(df["As"], df["Depth"], color='blue')

# Invert the y-axis so that depth increases downward
ax.invert_yaxis()

# Set x-axis labels on top
ax.xaxis.set_label_position('top')
ax.xaxis.tick_top()

# Set axis labels
ax.set_xlabel("Arsenic (ppb)", fontsize=12)
ax.set_ylabel("Depth (ft)", fontsize=12)

# Set x-limits starting at 0
ax.set_xlim(left=0)

# Display the plot
plt.show()
```

### Figure plotting arsenic concentrations (ppb) in water against well depths (ft) for wells in Bangladesh

## Question 3

```python
df2=pd.read_csv('well_data.csv')

# Define conditions for As concentration categories
low_arsenic = df2["As"] < 10
medium_arsenic = (df2["As"] >= 10) & (df2["As"] <= 50)
high_arsenic = df2["As"] > 50

# Plot each category with different colors and markers
fig, ax2 = plt.subplots(figsize=(8, 6))
ax2.scatter(df2["As"][low_arsenic], df2["Depth"][low_arsenic], 
           color='lightgreen', marker='o', label="As < 10")
ax2.scatter(df2["As"][medium_arsenic], df2["Depth"][medium_arsenic], 
           color='orange', marker='s', label="10  As  50")
ax2.scatter(df2["As"][high_arsenic], df2["Depth"][high_arsenic], 
           color='brown', marker='D', label="As > 50")

# Invert the y-axis so that depth increases downward
ax2.invert_yaxis()

# Set x-axis labels on top
ax2.xaxis.set_label_position('top')
ax2.xaxis.tick_top()

# Set axis labels
ax2.set_xlabel("Arsenic (ppb)", fontsize=12)
ax2.set_ylabel("Depth (ft)", fontsize=12)

# Set x-limits starting at 0
ax2.set_xlim(left=0)

# Add legend
ax2.legend(bbox_to_anchor=(0.98,0.2))

# Display the plot
plt.show()
```

### Figure plotting arsenic concentrations (ppb) in water against well depths (ft) for wells in Bangladesh separated out by color into low (<10), medium (10<=As<=50) and high (>50) concentrations
