```python
%matplotlib ipympl
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

## Questions 1-3

```python
df=pd.read_excel('Matthew_David_Ilomuanya_DataNB8.xlsx')
```

```python
df.describe()
```

```python
# setting x and y values
x=df['price (USD)']
y=df['marketcap']

# plot the data
fig,ax=plt.subplots()
fig.set_size_inches(6,6)
ax.scatter(x,y,label='Data')

ax.set_xlabel('Share Price (USD)',fontsize=16)
ax.set_ylabel('MarketCap (USD)',fontsize=16)

#### Do the best fit line and plot it
slope, intercept, r_value,p_value,stderr = stats.linregress(x,y)

x_fit=np.linspace(min(x),max(x))
y_fit=x_fit*slope+intercept
ax.plot(x_fit,y_fit,label='Fit')

# Use a comma in the x-axis for dollars
ax.xaxis.set_major_formatter('${x:,}')

# Title and display
title=('Figure Showing the Relationship Between Share Price \nand Market Cap of the top 50 Publicly Traded Companies')
ax.set_title(title)

props=dict(boxstyle='round',facecolor='wheat',alpha=0.5)

textstr='m={:.3f}\nb={:.3f}\n$r^2$={:.3f}\np={:.3f}'\
        .format(slope,intercept,r_value**2, p_value)
ax.text(0.60,0.95,textstr,transform=ax.transAxes\
        ,fontsize=10,verticalalignment='top',bbox=props)

plt.show()
```

### The p-value of 0.736 indicates no significant relationship between Share Price and MarketCap, suggesting the data aligns with the null hypothesis.
