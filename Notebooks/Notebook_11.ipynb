## Question 1

```python
%matplotlib ipympl
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
from scipy import stats
from matplotlib.backends.backend_pdf import PdfPages

# Load dataset from a CSV file
df = pd.read_csv('Bacteria_Arsenic_Data.csv')

# Create a figure and axis for plotting
fig, ax = plt.subplots()

# Define the variables for correlation analysis
x = 'hematiteppm'  # Independent variable: Hematite concentration in ppm
y = 'P-k__Bacteria;p__Acidobacteria;c__Acidobacteriia;o__Acidobacteriales'  # Dependent variable: Bacterial species abundance

# Define text box properties for displaying statistical results
props = dict(boxstyle='round', facecolor='coral', alpha=0.5)

# Print the selected bacterial species name
print(y)

# Check if the bacterial abundance data is of type float (numeric)
if df[y].dtype == float:
    # Create scatter plot of hematite concentration vs. bacterial abundance
    ax.scatter(df[x], df[y])

    # Set axis labels with larger font size for readability
    ax.set_xlabel("Hematite Concentration (ppm)", fontsize=14)
    ax.set_ylabel("Acidobacteriales (% Abundance)", fontsize=14)

    # Set the title of the plot
    plt.title(f"Correlation between Hematite and Acidobacteriales")

    # Perform linear regression analysis on the dataset, dropping NaN values
    results = stats.linregress(df[[x, y]].dropna())

    # Format statistical results into a text string for display
    textstr = 'm={:.3f}\nb={:.3f}\n$r^2$={:.3f}\np={:.3f}'.format(
        results.slope, results.intercept, results.rvalue ** 2, results.pvalue
    )

    # Generate x-values for the trend line (min and max values of hematite concentration)
    x_fit = np.array([df[x].min(), df[x].max()])

    # Plot the linear regression trend line in dotted gray
    ax.plot(x_fit, results.slope * x_fit + results.intercept, color='gray', linestyle='dotted')

    # Add the statistical results as a text box on the plot
    ax.text(0.75, 0.95, textstr, transform=ax.transAxes, fontsize=14,
            verticalalignment='top', bbox=props)

# Show the plot
plt.show()
```

### This is a figure showing the relationship between Hematite Concentration and Acidobacteriales (% Abundance) in sediment
### Acidobacteria are known for their abundance in soils, making up a significant portion of the bacterial community in various habitats. While their specific roles are still being researched, Acidobacteria are thought to play a role in nutrient cycling and other soil processes. 
### The data suggests that there is no significant correlation between the two with a p value of 0.547

## Plotting Top 4

```python
# Number of elements (bacterial species) to plot
el_to_plot = 4  

# Define number of rows and columns for subplots
nrows = 2  # Number of rows in the plot grid
ncols = int(el_to_plot / nrows)  # Number of columns (ensures a balanced layout)

# Create a figure and a 2D array of axes for subplots
fig, ax2d = plt.subplots(nrows, ncols)

# Set the figure size (width, height in inches)
fig.set_size_inches(10, 10)

# Flatten the 2D axes array into a 1D array for easier iteration
ax = np.ravel(ax2d)

# Adjust spacing between subplots to avoid overlapping
fig.subplots_adjust(wspace=0.4)

# Define text box properties for displaying statistical results
props = dict(boxstyle='round', facecolor='white', alpha=0.5)

# Loop through the top 'el_to_plot' bacterial species from df_r2['name']
for count, el in enumerate(df_r2['name'][0:el_to_plot]):
    x = 'hematiteppm'  # Independent variable: Hematite concentration
    y = el  # Dependent variable: Bacterial species abundance

    # Scatter plot of hematite concentration vs. bacterial species abundance
    ax[count].scatter(df[x], df[y])

    # Set axis labels with larger font size
    ax[count].set_xlabel(x, fontsize=14)
    ax[count].set_ylabel(y[-10:-1], fontsize=14)

    # Perform linear regression on the selected data (excluding NaNs)
    results = stats.linregress(df[[x, y]].dropna())

    # Format statistical results (slope, intercept, R², p-value) into a string
    textstr = 'm={:.3f}\nb={:.3f}\n$r^2$={:.3f}\np={:.3f}'.format(
        results.slope, results.intercept, results.rvalue ** 2, results.pvalue
    )

    # Generate x-values for the regression line (min and max values of hematite concentration)
    x_fit = np.array([df[x].min(), df[x].max()])

    # Plot the linear regression trend line in dotted gray
    ax[count].plot(x_fit, results.slope * x_fit + results.intercept, color='gray', linestyle='dotted')

    # Add the statistical results as a text box on the plot
    ax[count].text(0.7, 0.95, textstr, transform=ax[count].transAxes, fontsize=10,
                   verticalalignment='top', bbox=props)

# Show the final plot with all subplots
plt.show()
```
