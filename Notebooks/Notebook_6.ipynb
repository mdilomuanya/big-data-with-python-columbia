```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

## Hypothesis:
### The more coffee a student drinks, the higher their probability of finishing their homework on time

```python
# Creating dataset
coffee_intake = np.array([0, 1, 2, 3, 4, 5])  # Average cups of coffee consumed per day
homework_completion = np.array([20, 40, 65, 80, 90, 95])  # Percentage of homework completed on time
x = coffee_intake
y = homework_completion

# Manual Line Fitting
n = len(x)
m = (n*sum(x*y) - sum(x)*sum(y))/(n*sum(x**2) - sum(x)**2)
b = (sum(x**2)*sum(y) - (sum(x*y))*sum(x))/(n*sum(x**2) - sum(x)**2)

# Plotting manual graph
x_fit=np.linspace(min(x),max(x))
y_fit=m*x_fit+b

# Python's linregress method
slope, intercept, r_value, p_value, std_err = stats.linregress(coffee_intake, homework_completion)

# Plotting
fig,ax=plt.subplots()
ax.scatter(x,y,color='blue')
ax.plot(x_fit,y_fit,color='red', linestyle='-', label='Manual Fit')
ax.plot(x, slope * x + intercept, color='green', label='linregress Fit')
ax.set_xlabel('Average Cups of Coffee Consumed Per Day')
ax.set_ylabel('% of Homework Completed On Time')

# Title and display
title=('Figure Showing the Relationship Between Coffee Consumption \nand Homework Completion')
ax.set_title(title)

props=dict(boxstyle='round',facecolor='wheat',alpha=0.5)

textstr='m={:.3f}\nb={:.3f}\n$r^2$={:.3f}\np={:.3f}'\
        .format(slope,intercept,r_value**2, p_value)
ax.text(0.05,0.95,textstr,transform=ax.transAxes\
        ,fontsize=10,verticalalignment='top',bbox=props)

plt.show()
```

### This graph illustrates the relationship between coffee intake and homework completion both a manually computed line of best fit (dashed red) and Python's linregress fit (green) show a strong positive correlation, confirming our hypothesis

### The results suggest that coffee consumption has a significant positive correlation with homework completion and both methods of fitting the line give identical results, verifying the calculations
